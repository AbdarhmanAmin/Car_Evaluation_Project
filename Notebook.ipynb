# Declear Target Variable

```python
X=df.drop(["class"],axis=1)
y=df["class"]
```

```python
#split x and y into train and test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape, X_test.shape
```

# Feature Engineering 

```python
X_train.dtypes
```

Encoding Using Category Encoders

```python
import category_encoders as ce
```

```python
encoder=ce.OrdinalEncoder(cols=["buying","maint","doors","persons","lug_boot","safety"])
X_train=encoder.fit_transform(X_train)
X_test=encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_test.head()
```

 # Decision Tree Classifier using gini index


```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
gini=DecisionTreeClassifier(criterion='gini', max_depth=100,random_state=0)

gini.fit(X_train,y_train)
```

```python
y_pred_gini=gini.predict(X_test)
```

## Check accuracy score with gini index

```python
from sklearn.metrics import accuracy_score

print("Model accuracy score with gigi index: {0:0.4f}".format(accuracy_score(y_test,y_pred_gini)))
```

## Check OverFiting

```python
y_pred_train_gini=gini.predict(X_train)
```

```python
print("Model accuracy score with gigi index: {0:0.4f}".format(accuracy_score(y_train,y_pred_train_gini)))
```

```python
print("Training set score: {0:0.4f}".format(accuracy_score(y_train,y_pred_train_gini)))
print("Testing set score: {0:0.4f}".format(accuracy_score(y_test,y_pred_gini)))
```

 # Decision Tree Classifier using entropy


```python
entropy=DecisionTreeClassifier(criterion='entropy', max_depth=100,random_state=0)

entropy.fit(X_train,y_train)
```

```python
y_pred_entropy=entropy.predict(X_test)
```

## Check accuracy score with Entropy

```python
from sklearn.metrics import accuracy_score

print("Model accuracy score with Entropy: {0:0.4f}".format(accuracy_score(y_test,y_pred_entropy)))
```

## Check OverFiting

```python
y_pred_train_entropy=entropy.predict(X_train)
```

```python
print("Model accuracy score with gigi index: {0:0.4f}".format(accuracy_score(y_train,y_pred_train_gini)))
```

```python
print("Training set score: {0:0.4f}".format(accuracy_score(y_train,y_pred_train_entropy)))
print("Testing set score: {0:0.4f}".format(accuracy_score(y_test,y_pred_entropy)))
```

# Calculate Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_entropy)

print(cm)
```
